```python
#Panda is a Library used to perform table related operations
import pandas as pd
```

```python
# read csv data and assign into a variable dataset 
dataset = pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#get the input value from the dataset and assign it into the variable independent
independent = dataset[["YearsExperience"]]
independent
```

```python
#get the output value from the dataset and assign it into the variable dependent
dependent = dataset[["Salary"]]
dependent
```

```python
#split the train and test data where x_train = training data for input, x_test = testing data for input,
# y_train = training data for output,y_test = testing data for output
# sklearn is a library used for machine learning process
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Creating Model which means getting weight and bias by passing training input and training output
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#getting the value of weight
weight = regressor.coef_
weight
```

```python
#getting the value of bias(starting value)
bias = regressor.intercept_
bias
```

```python
#Evalution Metrics to cross check the test input data with predicted value
y_pred = regressor.predict(x_test)
```

```python
#Validating Parameter R Squared - To verify how well the created model is fitted - Actual test value Vs Predicted value
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```

```python
#if r2 value nearly to 1, the created model has better performance
r_score
```

```python
#Pickle is a library to read/write the file
import pickle
filename="finalized_model_linear.sav"
```

```python
#save the file into the location where wb means write binary
pickle.dump(regressor,open(filename,'wb'))
```

```python
#To verify the saved model is working as expected by passing test value.Where rb means read binary
loaded_model = pickle.load(open("finalized_model_linear.sav",'rb'))
result = loaded_model.predict([[15]])
```
