```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    #calculating VIF
    vif=pd.DataFrame()
    vif["Variables"]=X.columns
    vif["VIF"] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    return vif
```

Independent Sample - Unpaired T Test

Different Groups(Male,Female) but Same Condition(Salary)

```python
from scipy.stats import ttest_ind
dataset = dataset.dropna()
male = dataset[dataset["gender"] == "M"]["salary"]
female = dataset[dataset["gender"] == "F"]["salary"]
ttest_ind(male,female)
```

```python
dataset
```

```python
others = dataset[dataset["hsc_b"] == "Others"]["salary"]
central = dataset[dataset["hsc_b"] == "Central"]["salary"]
ttest_ind(others,central)
```

```python
# P-value of above condition is greater than 0.05%. So it tells that the groups(others and cetral board) students are different
```

Dependent Sample - Paired T Test

Same Groups(Male) but different Conditions(ssc_p,hsc_p)

```python
from scipy.stats import ttest_rel
dataset = dataset.dropna()
male = dataset[dataset["gender"] == "M"]["ssc_p"]
male1 = dataset[dataset["gender"] == "M"]["hsc_p"]
ttest_rel(male,male1)
```

```python
commerce = dataset[dataset["hsc_s"] == "Commerce"]["etest_p"]
commerce1 = dataset[dataset["hsc_s"] == "Commerce"]["mba_p"]
ttest_rel(commerce,commerce1)
```

```python
# P-value of above conditions is greater than 0.05%. So it tells that the conditions(etest_p and mba_p) marks are different
```
